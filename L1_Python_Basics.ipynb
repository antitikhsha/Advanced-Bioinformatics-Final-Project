**Types of Programming Paradigms:**
> Programming paradigms are a way to classify programming languages based on their features.
1.   Structured programming (C)

2.   Object-oriented programming (C++, Java)
        *   Programming paradigm that relies on the concept of classes and objects.   
3.   Functional programming
        *   Programs are constructed by applying and composing functions. (Haskell, Scala)

**Compiler and Interpreter Key Differences**
1.   Both are computer programs that transform code written in a high-level programming language into machine code. 
2.   Compiler scans the entire program and translates the whole of it into machine code at once.
3.   Interpreter translates just one statement of the program at a time into machine code.
4.   Compiled code runs faster while interpreted code runs slower.
5.   Compiler displays all errors after compilation, on the other hand, the Interpreter displays errors of each line one by one.

**Python Comments**
1.   Comments can be used to explain Python code.   
2.   Comments start with a #.
3.   Comments will not execute.
4.   Multiline comments are also possible with triple quote """.




```python
# This is a comment
print('How are you?')
```

```python
# Multi-line Comments
"""
Hello Students
How are you all?
"""
print("How are you?")
```
